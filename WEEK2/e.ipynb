# Data Visualization
<br> It is a very important task of each machine learning task.
<br> Python has several modules that we can use for data visualization.
<br> Among them the __[matplotlib](https://matplotlib.org/)__ is a very popular one. The ``` plt.scatter() ``` function is used to create scatterplots.

## Definition
<br> The scatterplot visualizes how two variables are related to each other.
<br> Each data point is represented as a single dot or marker of our preferred shape.
<br> We prefer the scatterplot visualization in cases where we wish to know the relationship between two variables.

## A first example
Suppose we have the weight in kilogramms for 6 people and their associated Body Mass Index (BMI) which is computed through the following formula : $BMI = Weight/ (Height) ^ 2$
<br> We wish to create a scatterplot to __understand the relationship between body weight and bmi value__.
<br> Firstly, we import the pyplot submodule from Matplotlib.
<br> Then, we create one list containing the body weight in kilograms and another list containing the associated bmi value.
<br> Afterwards, we create the scatterplot by using ```plt.scatter()``` with the two variables that we wish to compare as input arguments.
<br> Finally, we need to explicitly display the figure by using the ```plt.show()``` function.
<br> The ```plt.title(label,fontsize=NumericValue)``` is used to set a title for the entire plot.
<br> A similar syntax is followed for ```plt.xlabel()``` and ```plt.ylabel()``` for setting the titles for x and y axes respectively.
<br> The ```plt.grid(True,color='black',linestyle='-',linewidth=1)``` configures the grid lines.
<br> The ```plt.xticks(TicksArray, LabelsArray, fontsize=NumericValue)``` is used to set the current tick locations and labels on the x-axis.
<br> The syntax is similar for the ```plt.yticks()``` function.
<br> The ```numpy.arange(start,stop,step)``` returns evenly spaced values within a given interval.
<br> If we wish to use a __non-integer step__, such as the __0.1__, it is often better to use the ```numpy.linspace```function. It returns evenly spaced samples, calculated over the interval __StartValue...StopValue__.
<br>The code is the following one:

```python
import matplotlib.pyplot as plt

weight = [120, 50, 80, 65, 75, 92]
bmi = [34, 20, 28, 23, 26, 29]

plt.scatter(weight,bmi)
plt.title('BMI-Weight Association',fontsize=22)
plt.xlabel('Weight in kilograms',fontsize=16)
plt.ylabel('Body Mass Index',fontsize=16)
plt.xticks(np.arange(40, 121, 10), fontsize=12)
plt.yticks(np.arange(15, 36, 5), fontsize=12)
plt.grid(True)
plt.show()
```

# Data Distributions
1. A distribution is simply a collection of data or scores for a variable.
1. Usually, these scores are arranged in order from smallest to largest observations and then can be presented graphically.

# Distribution Terminology
1. Each data distribution is described by the associated density function which quantifies the proportion of data change over the range of the distribution.
1. There are two types of density functions.
1. The __Probability density function (PDF)__ calculates the probability of observing a given value.
1. The __Cumulative Density Function (CDF)__ calculates the probability of an observation equal or less than a value.

## Gaussian Distribution
1. It is also known as the __normal__ distribution.
1. This probability distribution is __symmetric__ about the mean value.
1. This imples that data __near the mean are more frequent__ in occurrence than data far from the mean.
1. It graphically appears as a "bell curve".
1. It can be described by just __two parameters__: The __mean value__ and the __standard deviation__.
1. For all normal distributions, __68.2% of the observations__ will appear within plus or minus __one standard deviation of the mean__.
1. __95.4% of the observations__ will fall within $+/-$ __two standard deviations of the mean__.
1. __99.7% of the observations__ will fall within $+/-$ __three standard deviations of the mean__.
<br>
<br>
The following code segment is used to generate a normal distribution and plot its probability density function:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import gaussian_kde

# Set a random seed for reproducibility
np.random.seed(42)

#Generate synthetic data from a normal distribution
mean_value =3.0
standard_deviation = 0.5
sample_size = 1000
synthetic_data = np.random.normal (mean_value, standard_deviation, sample_size)

# Plot the histogram of the synthetic data
plt.hist(synthetic_data, bins =30, density=True, alpha=0.5, label="Histogram")

# Fit a KDE to the data
kde = gaussian_kde(synthetic_data)
x_vals = np.linspace(min(synthetic_data),max(synthetic_data), 1000)
kde_vals = kde(x_vals)

# Plot the estimated PDF using KDE
plt.plot(x_vals, kde_vals, label="KDE", color='red')

# Plot the true PDF of the normal distribution for comparison
pdf_vals = norm.pdf(x_vals, mean_value, standard_deviation)
plt.plot(x_vals,pdf_vals, label="True PDF", linestyle='dashed', color='green')

# Set labels and legend
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.legend()

# Show the plot
plt.show()
```

## Uniform Distribution
1. A uniform distribution is a probability distribution type in which all the values within a given range have the same possibility to occur.
1. The ```numpy.random.uniform(LowValue, HighValue, Size)``` draws samples from a uniform distribution.
1. It models situations where each outcome has an equal chance to happen.
1. So, there is no bias or preference for any particual value within the range.
1. The range of the distribution is defined by its lower bound (a) and its upper bound (b)
1. Its probability density function is given by: $f(x) =
\begin{cases}
    1 / b-a, & \text{if } a \leq x \leq b \\
    0 , & \text{if }  x < a \\
    0, & \text{if } x > b
\end{cases}
$

The following code segment is used to generate a uniform distribution between 2 and 6, while also plotting its probability density function (PDF):
``` python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of the uniform distribution
a = 2 # Lower bound
b = 6 # Upper bound

# Generate synthetic data from a uniform distribution
sample_size = 5000
uniform_data = np. random.uniform(a,b,sample_size)

# Plot the histogram of the synthetic data to visualize its distribution
plt.hist(uniform_data, bins = 30, density=True, alpha=0.5, label="Histogram")

#Plot the PDF of the uniform distribution
x_vals=np.linspace(a-1,b+1,sample_size) # Create a range of x values
pdf_vals = np.full_like(x_vals, 1/(b-a)) # PDF is constant within [a,b]
plt.plot(x_vals, pdf_vals, label='PDF',color='red')

# Set labels and legend
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.legend()

#Show the plot
plt.show()
```

## Exercise 2: Find the decision boundary

**Find a correct decision boundary** that separates the classes created during the previous task. That decision boundary is defined with a function called `decision_boundary`. In this case, the decision boundary is a line. The general equation of a line is:
$$
y = ax + b
$$
Where the parameter $a$ is called `slope` and defines how steep the line is, and the parameter $b$ is called `intercept` and defines the point where the line intersects the y-axis (for $x=0$).

**Change the values of** `slope` and `intercept` and **plot** the function defined by `decision_boundary` on the interval $x=[-7,7]$ in green.

_Note:_ As you change the `slope` and `intercept` values, the provided code will plot your samples classified according to the decsion boundary you have defined.

You should be using `np.linspace` and `np.plot` to solve this task.

````python
import numpy as np
import matplotlib.pyplot as plt

# CHANGE THE VALUES OF SLOPE AND INTERCEPT
def decision_boundary(x):
    slope = 1.2
    intercept = 1
    return slope*x + intercept


## FILL OUT TO PLOT THE DECISION BOUNDARY:
# Create Classes - place the code here
import numpy as np
import matplotlib.pyplot as plt

sample = 100
np.random.seed(0) #fixes seed so is not random
x1 = np.random.uniform(-2,2,sample) #boundaries and sample size
y1 = np.random.uniform(4,7,sample)

x2 = np.random.uniform(3,7,sample)
y2 = np.random.uniform(0,4,sample)

#Plot the decision boundary.
x_values = np.linspace(-7, 7, 100)
y_values = decision_boundary(x_values)


plt.scatter(x1,y1,label = "Class 1", c="blue")
plt.scatter(x2,y2,label = "Class 2", c="red")
plt.plot(x_values, y_values, label='Decision Boundary', color='green') #plotting the boundary.
plt.grid(True)
plt.legend()
plt.show()


````

# An introduction to regression analysis
1. It is a common machine learning task to estimate the relationship between the __outcome__ of a given task __(dependnet variable)__ and one or more variables that we use as __features__ to __predict__that outcome (__independnet variables)__.
1. The main difference between classification and regression is the __type of the output variable__.
1. We try to predict a __categorical or discrete output__ when dealing with a __classification__ problem.
1. In __regression__, the __output is numerical or continuous__.
1. The __linear regression__ is the most common model in regression analysis.
1. This model finds the relationship between the independent and dependent variables by __fitting a linear equation__.
1. The f(x) or y-value represents the independent variable, which we wish to predict.
1. The x-value, which is usually in the horizontal axis, shows the independent variable values.
1. We are making predictions by entering the independent variable's value (x-value) in the regression equation, and then computing the value of the idependent variable.

### Description of a Piecewise Function

Let's define a regression line through a piecewise function \(y\) that takes different forms in different intervals:


$f(x) =
\begin{cases}
    x^3, & \text{if } x \leq 2 \\
    2x + 4, & \text{if } 2 < x \leq 10 \\
    0.2x^2 + 4, & \text{if } x > 10
\end{cases}
$

- When \(x\) is less than or equal to 2, the function is defined as \(f(x) = x^3\).
- When \(x\) is greater than 2 but less than or equal to 10, the function is defined as \(f(x) = 2x + 4\).
- When \(x\) is greater than 10, the function is defined as \(f(x) = 0.2x^2 + 4\).


### Regression line code

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of the x values
x=np.linspace(0,20,400)

# Initialize an empty array for y values
y=np.empty_like(x)

# Calculate y values based on the piecewise function
for i, xi in enumerate(x):
    if xi<= 2:
        y[i]= pow(xi,3)
    elif xi> 2 and xi <=10:
        y[i] = 2*xi+4
    else:
        y[i] = 0.2*xi * xi +21.2

# Plot the function
plt.figure(figsize=(8,6))
plt.plot(x,y,label='Regression Line')
plt.xlim(0,20)
plt.ylim(0,84)
plt.xticks(np.arange(0, 20, 1), fontsize=12)
plt.yticks(np.arange(0, 84, 4), fontsize=12)
plt.xlabel('x-value', fontsize=22)
plt.ylabel('y-value',fontsize =22)
plt.legend(fontsize=24)
plt.grid(True)
plt.show()
````
